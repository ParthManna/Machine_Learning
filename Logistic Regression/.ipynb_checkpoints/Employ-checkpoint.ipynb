```python
import pandas as pd
```

```python
df = pd.read_csv('CSV_file/HR_comma_sep.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visual styling
sns.set(style="whitegrid")

# Plot the impact of salary on employee retention
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='salary', hue='left', palette='Set1')
plt.title('Impact of Salary on Employee Retention')
plt.xlabel('Salary Category')
plt.ylabel('Number of Employees')
plt.legend(title='Left the Company', loc='upper right')
plt.show()
```

```python

# Plot the correlation between department and employee retention
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Department', hue='left', palette='Set2')
plt.title('Correlation Between Department and Employee Retention')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
# plt.xticks(rotation=45)
plt.legend(title='Left the Company', loc='upper right')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python

# Prepare the data for modeling

# Convert categorical variables into dummy variables (one-hot encoding)
df_encoded = pd.get_dummies(df, columns=['salary', 'Department'], drop_first=True).astype(int)
print(df_encoded)
```

```python

# Features and target variable
X = df_encoded.drop('left', axis=1)
y = df_encoded['left']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
X_test_scaled
```

```python

# Build the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)
```

```python

# Measure the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
